```python
# importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

#headers for requests
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
           "Accept-Language": "en-US, en;q=0.5"}

base_url = "https://www.amazon.com"
url = "https://www.amazon.com/s?k=laptop&crid=21S8OXJT0EGY9&sprefix=laptop%2Caps%2C483&ref=nb_sb_noss_"

#creating lists to store the data
product_names = []
product_prices = []
product_ratings = []
product_urls = []

for page in range(1, 20):
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "lxml")
    
    # Extract product details
    titles = soup.find_all("span", class_="a-size-medium a-color-base a-text-normal")
    prices = soup.find_all("span", class_="a-price-whole")
    ratings = soup.find_all("span", class_="a-icon-alt")
    links = soup.find_all("a", class_="a-link-normal s-no-outline")
    
    # Ensure the same number of items in each category
    for i in range(len(titles)):
        product_names.append(titles[i].text.strip() if i < len(titles) else None)
        product_prices.append(prices[i].text.strip() if i < len(prices) else None)
        product_ratings.append(ratings[i].text.strip() if i < len(ratings) else None)
        product_urls.append(base_url + links[i].get('href') if i < len(links) else None)
    
    # Find the next page link
    next_page = soup.find("a", class_="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator")
    
    if next_page:
        next_page_url = next_page.get("href")
        url = base_url + next_page_url
        print(f"Scraping page {page + 1}: {url}")
    else:
        print("No more pages found.")
        break
    
    # Add delay to prevent being blocked by Amazon
    time.sleep(2)

# Create DataFrame
df = pd.DataFrame({
    "Product_Name": product_names,
    "Price": product_prices,
    "Rating": product_ratings,
    "Product_URL": product_urls
})

# Save to CSV
df.to_csv("amazon_product_data.csv", index=False)
```
